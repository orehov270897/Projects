```python
import pandas as pd
bookings = pd.read_csv('bookings.csv', sep=';') # чтение файла .csv с данными о бронированиях отелей
bookings_head = bookings[:7]
```

```python
for col_name in bookings.columns:
    # приведение названий столбцов к нижнему регистру и замена ' ' на '_'
    bookings = bookings.rename(columns={col_name: col_name.lower().replace(' ', '_')})
```

```python
bookings
```

```python
bookings.query('is_canceled == 0').groupby('country').agg({'hotel': 'count'}).sort_values('hotel', ascending=False)
# подсчет количества успешных(is_canceled == 0) бронирований по странам
```

```python
bookings.groupby('hotel').agg({'stays_total_nights': 'mean'}).round(2)
```

```python
# посчет наблюдений овербукинга
bookings.query('reserved_room_type != assigned_room_type').shape
```

```python
# посчет количества заселений по месяцам
bookings.groupby(['arrival_date_year', 'arrival_date_month']) \
        .agg({'arrival_date_day_of_month': 'count'})#.sort_values('arrival_date_day_of_month')
```

```python
# посчет количества отмененных заселений по месяцам для City Hotel
bookings.query('is_canceled == 1' and "hotel == 'City Hotel'") \
        .groupby(['arrival_date_year', 'arrival_date_month']).agg({'arrival_date_day_of_month': 'count'})
```

```python
bookings.agg({'adults': 'mean', 'children': 'mean', 'babies': 'mean'})
```

```python
bookings['total_kids'] = bookings['children'] + bookings['babies']
```

```python
bookings.groupby('hotel')['total_kids'].mean().round(2)
```

```python
bookings['has_kids'] = (bookings['total_kids'] != 0)
```

```python
bookings
```

```python
p_with_c = round((bookings.query('has_kids == True and is_canceled == 1').agg({'is_canceled': 'count'}) / 
            bookings.query('has_kids == True').agg({'is_canceled': 'count'})) * 100, 2)
p_without_c = round((bookings.query('has_kids == False and is_canceled == 1').agg({'is_canceled': 'count'}) / 
            bookings.query('has_kids == False').agg({'is_canceled': 'count'})) * 100, 2)
```
