```python
import pandas as pd
import numpy as np
```

```python
datafile = open('gff.tsv')
data = datafile.readlines()
data[2].split('	')
```

```python
def readstr(s):
    s = s.split('	')
    last_cell = s[len(s)-1]
    last_cell = last_cell.split(';')[-1].split('=')[1]
    last_cell = last_cell[:-1]
    if last_cell[-1] == ',':
        last_cell = last_cell[:-1]
    s[len(s)-1] = last_cell
    return s
```

```python
data_df = pd.DataFrame([readstr(i) for i in data[1:]], columns=data[0].split('	'))
data_df
```

```python
data_df.groupby('type').agg({'type': 'count'})
```
