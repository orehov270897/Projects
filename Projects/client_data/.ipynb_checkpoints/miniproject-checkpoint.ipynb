```python
user_data = pd.read_csv('3_user_data.csv')
```

```python
logs = pd.read_csv('3_logs.csv')
```

```python
logs
```

```python
logs.platform.nunique()
```

```python
logs.query('success == True').groupby('client', as_index = False).agg({'success': 'count'}).query('success == 41') \
    .sort_values('client', ascending = True)
```

```python
logs.query('success == True').groupby('platform', as_index = False).agg({'success': 'count'})
```

```python
data = logs.merge(user_data, how='inner', on='client')
```

```python
data
```

```python
data.query('premium == True').groupby('platform').agg({'client': 'count'})
```

```python
sns.distplot(data.query('premium == True').age)
sns.distplot(data.query('premium == False').age)
```
