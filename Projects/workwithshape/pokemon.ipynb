```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pokemon = pd.read_csv('Pokemon.csv')
pokemon
```

```python
pokemon = pokemon.rename(columns={'#': 'id'})
pokemon.columns = [i.lower().replace('.', '').replace(' ', '_') for i in pokemon.columns]
pokemon
```

```python
legends = pokemon.groupby('generation').legendary.value_counts().to_frame()
legends
```

```python
legends.rename(columns={'legendary': 'legendary_count'}, inplace=True)
legends
```

```python
legends_unstacked = legends.unstack(level=1)
legends_unstacked
```

```python
pokemon.groupby(['generation', 'type_1']).legendary.value_counts().to_frame().unstack(level=1)
```

```python
avocado_agg = pd.DataFrame({'type' : ['conventional', 'organic'],
                            'AvgPrice_2015' : [1.077963, 1.673324],
                            'AvgPrice_2016' : [1.105595, 1.571684],
                            'AvgPrice_2017' : [1.294888, 1.735521],
                            'AvgPrice_2018' : [1.127886, 1.567176],
                            })
avocado_agg
```

```python
avocado_agg_long = pd.wide_to_long(avocado_agg, 'AvgPrice', i='type', j='year', sep='_')
```
