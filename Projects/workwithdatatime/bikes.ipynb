```python
bikes_april = pd.read_csv('bikes_april.csv')
bikes_april
```

```python
bikes_april.start_time = pd.to_datetime(bikes_april.start_time)
```

```python
bikes_april.set_index('start_time', inplace=True) # Установка времени старта в качестве индекса
```

```python
bikes_april_new = bikes_april.resample(rule='D').usertype.value_counts().to_frame() \
                             .rename(columns={'usertype': 'quantity'}).unstack(level=1)
bikes_april_new
```

```python
bikes_april_new = bikes_april_new.droplevel(level=0, axis=1) # Удаление 0-го уровня названия колонок ("quantity")
```

```python
bikes_april_new.columns = bikes_april_new.columns.rename('') # Удаление имени ряда названий колонок("usertype")
```

```python
bikes_april_new = bikes_april_new.reset_index()
```

```python
def getdf(filenamelist):
    df = pd.read_csv(filenamelist[0])
    for i in range(1, len(filenamelist)-1):
        df_cur = pd.read_csv(filenamelist[i])
        df = pd.concat([df, df_cur])
    return df
```

```python
bikes_general = getdf(['bikes_q2_sample_apr.csv', 'bikes_q2_sample_may.csv', 'bikes_q2_sample_jun.csv',
                       'bikes_q3_sample_july.csv', 'bikes_q3_sample_aug.csv', 'bikes_q3_sample_sep.csv',
                       'bikes_q4_sample_oct.csv', 'bikes_q4_sample_nov.csv', 'bikes_q4_sample_nov.csv'])
```

```python
bikes_general
```

```python
bikes_general.start_time = pd.to_datetime(bikes_general.start_time) 
```

```python
bikes_general.set_index('start_time', inplace=True) # Установка времени старта в качестве индекса
```

```python
bikes_general_new = bikes_general.resample(rule='D').usertype.value_counts().to_frame() \
                             .rename(columns={'usertype': 'quantity'}).unstack(level=1) \
                             .droplevel(level=0, axis=1) # Удаление 0-го уровня названия колонок ("quantity")
bikes_general_new
```

```python
bikes_general_new.columns = bikes_general_new.columns.rename('') # Удаление имени ряда названий колонок("usertype")
```

```python
bikes_general_new[bikes_general_new.Customer > bikes_general_new.Subscriber]
```

```python
bikes_summer = bikes_general.loc['2018-06-01':'2018-08-31']
bikes_summer
```

```python
top_destination = bikes_summer.to_station_name.value_counts().idxmax() # Определение наиболее популярного пункта назначения
top_destination
```

```python
bad_day = bikes_summer[bikes_summer.to_station_name == top_destination].resample(rule='D').to_station_name.count().idxmin() \
                                                             .strftime('%Y-%m-%d')
bad_day
```
