```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
actlogs = pd.read_csv('prod_activations_logs.csv', sep=';')
actlogs.reset_index()
```

```python
tm_sales = []
for i in range(3):
    tm = pd.read_csv('tm_sales_' + str(i + 1) + '.csv', sep=';')
    tm_sales.append(tm)
tm_sales[2]
```

```python
tm = pd.concat([tm_sales[0], tm_sales[1], tm_sales[2]])
```

```python
tm
```

```python
def append_id(string):
    if str(string)[:2] != 'id':
        string = 'id' + str(string)
    return string
tm.SUBS_ID = tm.SUBS_ID.apply(append_id)
```

```python
tm = tm.reset_index()
tm
```

```python
tm = tm.drop(tm[tm.SUBS_ID == 'idnan'].index)
tm
```

```python
tm = tm.merge(actlogs, how='inner', on=['SUBS_ID', 'PROD_ID'])
```

```python
tm
```

```python
tm.ACT_DTTM = pd.to_datetime(tm.ACT_DTTM, format='%d-%m-%Y %H:%M')
tm.START_DTTM = pd.to_datetime(tm.START_DTTM, format='%d-%m-%Y %H:%M')
tm.END_DTTM = pd.to_datetime(tm.END_DTTM, format='%d-%m-%Y %H:%M')
```

```python
tm['min'] = tm.END_DTTM - tm.START_DTTM
tm
```
